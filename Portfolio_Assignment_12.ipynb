```python
import pandas as pd
import seaborn as sns
pokemon = pd.read_csv('pokedex_(Update_04.21).csv', sep=',')
pd.set_option('max_columns', 100)
```

```python
pokemon.head()
```

```python
typeMeans = pokemon.groupby("type_1").mean()
```

```python
#Er is met 99% zekerheid te zeggen dat Normal pokemon een lagere attack hebben dan Sub Legendaries, Legendaries en Mythical pokemon.
sns.barplot(ci=99, y="attack", x="status", data=pokemon)
```

```python
#Er is met 95% Zekerheid te zeggen dat Legendary pokemon hogere total_points hebben dan Normal, Sub Legendary en Mythical pokemon.
sns.barplot(ci=95, y="total_points", x="status", data=pokemon)
```

```python
#Er is met 99.99% zekerheid te zeggen dat Normal pokemon makkelijker te vangen zijn dan Sub Legendary, Legendary, en Mythical pokemon.
sns.barplot(ci=95, y="catch_rate", x="status", data=pokemon)
```
