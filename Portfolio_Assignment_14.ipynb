```python
import pandas as pd
import seaborn as sns
pokemon = pd.read_csv('pokedex_(Update_04.21).csv', sep=',')
pd.set_option('max_columns', 100)
```

```python
pokemon
```

```python
contingencyTable = pokemon.groupby(['type_number','generation']).size().unstack('type_number', fill_value=0)
contingencyTable
```

```python
contingencyTable.plot(kind="bar")
```

Er zijn geen grote verschillen in aantal types per generatie, volgens de contingency table


Er zijn geen grote verschillen in aantal types per generatie, volgens de bar plot

```python
from scipy.stats import chi2_contingency
chi2_contingency(contingencyTable)
```

De kans is 11.8% dat er geen correlatie is.
