```python
import pandas as pd
```

```python
pokemon = pd.read_csv('pokedex_(Update_04.21).csv', sep=',')
pd.set_option('max_columns', 100)
```

```python
# First types of all pokemon
pokemon['type_1'].value_counts(dropna=False).plot(kind='bar')
```

```python
# second type of all pokemon
pokemon['type_2'].value_counts().plot(kind='bar')
```

```python
#Defense stats of pokemon
pokemon['defense'].plot(kind='hist', bins=10)
```

```python
#attack stats of pokemon
pokemon['attack'].plot(kind='box')
```
